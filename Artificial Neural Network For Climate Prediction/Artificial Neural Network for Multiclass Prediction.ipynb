```python
x=df.iloc[:,:-1].to_numpy()
y=df["event_type"].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=19)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
X_train = ss.fit_transform(x_train)
X_test = ss.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout,Input
from tensorflow.keras.callbacks import EarlyStopping
```

```python
def create_ann(input_dim, num_classes):
    model = Sequential()
    model.add(Input(shape=(input_dim,)))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
early_stop = EarlyStopping(
    monitor='val_loss',  
    patience=10,
    restore_best_weights=True
)
```

```python
from scikeras.wrappers import KerasClassifier
```

```python
ann = KerasClassifier(
    model=create_ann,
    model__input_dim=x_train.shape[1],
    model__num_classes=len(np.unique(y_train)),
    epochs=100,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("ann", ann)
])
```

```python
pipe.fit(x_train, y_train)
```

```python
keras_model = pipe.named_steps['ann'].model_
keras_model.summary()
```

```python
y_pred = pipe.predict(x_test)
y_pred_prob = pipe.predict_proba(x_test)
```

```python
df["event_type"] = "Normal"


# FLOOD (balanced threshold)

df.loc[
    (df["precipitation"] >= 30) &     # Lowered for your dataset
    (df["soil_moisture"] >= 25) &
    (df["cloud_cover"] >= 50) &
    (df["humidity"] >= 55) &
    (df["storm_severity"] >= 2),      # Very achievable
    "event_type"
] = "Flood"


# HEATWAVE

df.loc[
    (df["event_type"] == "Normal") &
    (df["temp_max"] >= 40) &
    (df["temp_avg"] >= 32) &
    (df["humidity"] <= 35) &
    (df["solar_rad"] >= 450) &
    (df["uv_index"] >= 7),
    "event_type"
] = "Heatwave"


# SMOG

df.loc[
    (df["event_type"] == "Normal") &
    (df["aqi"] >= 150) &
    (df["pm25"] >= 90) &
    (df["pm10"] >= 80) &
    (df["no2"] >= 50) &
    (df["so2"] >= 40) &
    (df["co_gas"] >= 1.0),
    "event_type"
] = "Smog"


# WIND STORM

df.loc[
    (df["event_type"] == "Normal") &
    (df["wind_speed"] >= 18) &
    (df["storm_severity"] >= 3) &
    (df["pressure"] <= 1008),
    "event_type"
] = "Wind Storm"


# EARTHQUAKE RISK

df.loc[
    (df["event_type"] == "Normal") &
    (
        (df["fault_activity"] >= 6) |
        (df["micro_tremors"] >= 3)
    ) &
    (df["tectonic_pressure"] >= 40) &
    (df["earth_magnetic_field"] >= 48),
    "event_type"
] = "Earthquake Risk"


# DROUGHT

df.loc[
    (df["event_type"] == "Normal") &
    (df["precipitation"] <= 2) &
    (df["humidity"] <= 30) &
    (df["soil_moisture"] <= 20) &
    (df["temp_max"] >= 36),
    "event_type"
] = "Drought"


# RAIN

df.loc[
    (df["event_type"] == "Normal") &
    (df["precipitation"] > 5) &
    (df["precipitation"] <= 60) &
    (df["cloud_cover"] >= 50) &
    (df["humidity"] >= 40),
    "event_type"
] = "Rain"


# FOREST FIRE (NEW)

df.loc[
    (df["event_type"] == "Normal") &
    (df["temp_max"] >= 38) &
    (df["humidity"] <= 30) &
    (df["soil_moisture"] <= 25) &
    (df["fire_risk_index"] >= 50) &   # achievable
    (df["visibility"] <= 8),
    "event_type"
] = "Forest Fire"
```
